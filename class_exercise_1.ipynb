```python
user_name = ""
account_balance = 0.0

def create_account(name, initial_balance=0.0):
    global user_name, account_balance
    user_name = name
    account_balance = initial_balance
    print(f"Account created for {user_name} with initial balance: {account_balance:.2f}")

def deposit(amount):
    global account_balance
    if amount > 0:
        account_balance += amount
        print(f"Deposit successful. New balance: {account_balance:.2f}")
    else:
        print("Invalid deposit amount.")

def withdraw(amount):
    global account_balance
    if amount > 0 and account_balance >= amount:
        account_balance -= amount
        print(f"Withdrawal successful. New balance: {account_balance:.2f}")
    elif amount <= 0:
        print("Invalid withdrawal amount.")
    else:
        print("Insufficient balance.")

def check_balance():
    global user_name, account_balance
    print(f"Account Holder: {user_name}")
    print(f"Current Balance: {account_balance:.2f}")

def main():
    print("--- Simple Banking System ---")
    create_account("John Doe", 100.00)

    while True:
        print("\nMenu:")
        print("1. Deposit")
        print("2. Withdraw")
        print("3. Check Balance")
        print("4. Exit")
        choice = input("Enter your choice: ")

        if choice == '1':
                amount = float(input("Enter amount to deposit: "))
                deposit(amount)

        elif choice == '2':
                amount = float(input("Enter amount to withdraw: "))
                withdraw(amount)


        elif choice == '3':
            check_balance()
        elif choice == '4':
            print("Exiting banking system. Goodbye!")
            break
        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```

```python
import math

class Shape:
    def __init__(self, length, width=2):
        self.length = length
        self.width = width

    def get_area(self):
        return 0

    def display(self):
        print(f"Shape with length {self.length}, width {self.width}")


class Rectangle(Shape):
    def __init__(self, length, width):
        super().__init__(length, width)

    def get_area(self):
        return self.length * self.width

    def display(self):
        print(f"Rectangle: Area {self.get_area()}")


class Circle(Shape):
    def __init__(self, radius):
        super().__init__(radius, 0)
        self.radius = radius

    def get_area(self):
        return math.pi * self.radius * self.radius

    def display(self):
        print(f"Circle: Radius {self.radius}, Area {self.get_area()}")



def main():
    rect = Rectangle(5, 4)
    circ = Circle(3)

    rect.display()
    circ.display()



if __name__ == "__main__":
    main()
```

```python
class BankAccount:

    def __init__(self, name, balance=2000):
        self.account_holder = name
        self.balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Deposited: {amount}")
        else:
            print("Invalid deposit amount!")

    def withdraw(self, amount):
        if amount > 0 and amount <= self.balance:
            self.balance -= amount
            print(f"Withdrawn: {amount}")
        else:
            print("Insufficient balance or invalid amount!")

    def display(self):
        print("Account Holder:", self.account_holder)
        print("Balance:", self.balance)


def main():
    account = BankAccount("John", 5000)

    account.display()

    account.deposit(1500)

    account.withdraw(2000)

    account.display()



if __name__ == "__main__":
    main()
```

```python
class Member:
    def _init_(self, name, membership_id):
        self.name = name
        self.membership_id = membership_id

    def display(self):
        pass


class RegularMember(Member):
    def _init_(self, name, membership_id):
        super()._init_(name, membership_id)
        self.borrow_limit = 3

    def getBorrowLimit(self):
        return self.borrow_limit

    def display(self):
        print("Member Type : Regular Member")
        print("Name        :", self.name)
        print("Borrow Limit:", self.getBorrowLimit())
        print()



class PremiumMember(Member):
    def _init_(self, name, membership_id):
        super()._init_(name, membership_id)
        self.borrow_limit = 10
        self.discount = "Available"

    def getBorrowLimit(self):
        return self.borrow_limit

    def display(self):
        print("Member Type : Premium Member")
        print("Name        :", self.name)
        print("Borrow Limit:", self.getBorrowLimit())
        print("Discount    :", self.discount)
        print()


def main():
    regular = RegularMember("Arun", "RM101")
    premium = PremiumMember("Priya", "PM201")

    regular.display()
    premium.display()


if __name__ == "_main_":
    main()
```
